```python

import json
import os
import subprocess
import csv
import numpy as np
from tqdm import tqdm
from datetime import datetime
from nltk.translate.bleu_score import corpus_bleu, SmoothingFunction

RPCD_DIR   = '/content/Aestheval/data/reddit/RPCD/'
IMAGES_DIR = '/content/Aestheval/data/reddit/RPCD/photocritique/images'  # base path for im_paths

with open(f'{RPCD_DIR}/reddit_photocritique_image_comments.json') as f:
    data = json.load(f)

im_paths  = data['im_paths']                      # list of image paths
comments  = data['first_level_comments_values']   # list of list of comments

print(f"Images:  {len(im_paths):,}")
print(f"Comment groups: {len(comments):,}")
assert len(im_paths) == len(comments), "Mismatch!"

import pandas as pd
rpcd_df = pd.DataFrame({
    'img_path':  [os.path.join(IMAGES_DIR, p) for p in im_paths],
    'comments':  comments,
})

rpcd_df = rpcd_df[
    rpcd_df['img_path'].apply(os.path.exists) &
    rpcd_df['comments'].apply(lambda c: len(c) > 0)
].reset_index(drop=True)

print(f"Valid pairs: {len(rpcd_df):,}")
print(f"\nSample image: {rpcd_df.iloc[0]['img_path']}")
print(f"Sample comment: {rpcd_df.iloc[0]['comments'][0][:100]}...")
```

```python
GOOGLE_RESEARCH = '/content/google-research'
CKPT_VILAR      = '/content/checkpoints/vila_rank_tuned'
CKPT_VILAP      = '/content/checkpoints/vila/vila_pretrain'
SPM_PATH        = '/content/checkpoints/vila/spm_model/spm.model'
VILA_PYTHON = '/usr/local/envs/vila_env/bin/python'

def generate_caption(image_path):
    """Generate aesthetic caption using VILA-P via CLI."""
    result = subprocess.run(
        [
            VILA_PYTHON, '-m', 'vila.run_vila_decode',
            '--is_pretrain',
            f'--ckpt_dir={CKPT_VILAP}',
            f'--image_path={image_path}',
            f'--spm_model_path={SPM_PATH}',
        ],
        capture_output=True,
        text=True,
        cwd=GOOGLE_RESEARCH,
    )

    for line in result.stdout.splitlines():
        if 'VILA generated comment' in line:
            raw = line.split(':', 1)[-1].strip()
            caption = raw.strip("[]'\" ")
            return caption

    return lines[-1] if lines else None
```

```python
# ============================================================
# VILA-R inference przez subprocess (CLI)
# ============================================================
import subprocess, os, re
import numpy as np
import pandas as pd
from tqdm import tqdm
from scipy.stats import spearmanr, pearsonr
GOOGLE_RESEARCH = '/content/google-research'
CKPT_VILAR = '/content/checkpoints/vila/vila_rank_tuned'
CKPT_VILAP = '/content/checkpoints/vila/vila_pretrain'
SPM_PATH   = '/content/checkpoints/vila/spm_model/spm.model'
VILA_PYTHON = '/usr/local/envs/vila_env/bin/python'

def run_vila_predict(image_path, ckpt_dir, spm_path):


    result_QC = subprocess.run(
        [
            VILA_PYTHON, '-m', 'vila.run_vila_predict',
            f'--ckpt_dir={ckpt_dir}',
            f'--image_path={image_path}',
            f'--spm_model_path={spm_path}',
        ],
        capture_output=True,
        text=True,
        cwd=GOOGLE_RESEARCH,
    )
    quality_score, zsl_single_score, zsl_ensemble_score = None, None, None
    if result_QC.returncode != 0:
        print(f"[ERROR]: {image_path}:\n{result_QC.stderr}")

    for line in result_QC.stdout.splitlines():
        # Linia: "===== VILA predicted quality score [0, 1]:  [[0.73]]"
        if 'VILA predicted quality score' in line:
            nums = re.findall(r'[\d.]+', line.split(':')[-1])
            if nums:
                quality_score = float(nums[0])
        if 'VILA ZSL single prompt score' in line:
            nums = re.findall(r'[\d.]+', line.split(':')[-1])
            if nums:
                zsl_single_score = float(nums[0])
        if 'VILA ZSL ensemble score' in line:
            nums = re.findall(r'[\d.]+', line.split(':')[-1])
            if nums:
                zsl_ensemble_score = float(nums[0])

    if quality_score is None or zsl_single_score is None or zsl_ensemble_score is None:
        print(f"[ERROR]: Not all scores were calculated")
        return None, None, None
    return quality_score, zsl_single_score, zsl_ensemble_score
```

```python
import csv
import os
from tqdm.auto import tqdm
from datetime import datetime

LIMIT = 1000  # Set to None to evaluate the full test set
VILA_PYTHON = '/usr/local/envs/vila_env/bin/python'
GOOGLE_RESEARCH = '/content/google-research'

# --- OUTPUT FILE SETUP ---
RESULTS_DIR = '/content/drive/MyDrive/EEML_VILA/results'
os.makedirs(RESULTS_DIR, exist_ok=True)

ts       = datetime.now().strftime('%Y%m%d_%H%M')
csv_path = os.path.join(RESULTS_DIR, f'predictions_{ts}.csv')

# Write CSV header immediately — so partial results are saved even if crash
with open(csv_path, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['img_path', 'zsl_single', 'zsl_ensemble', 'vila_rank', 'mos_gt'])
print(f"Results will be saved to: {csv_path}")

# --- DATA PREPARATION ---
df_eval = test_df.iloc[:LIMIT] if LIMIT else test_df

vilar_scores    = []
zsl_single_scores = []
zsl_ensemble_scores = []
gt_mos          = []

# --- EVALUATION LOOP ---
for _, row in tqdm(df_eval.iterrows(), total=len(df_eval), desc='VILA Evaluation'):
    # Call the inference function which handles subprocess.run internally
    q, zsl_s, zsl_e = run_vila_predict(row['img_path'], CKPT_VILAR, SPM_PATH)

    if q is not None:
        vilar_scores.append(q)
        zsl_single_scores.append(zsl_s   if zsl_s is not None else 0.0)
        zsl_ensemble_scores.append(zsl_e if zsl_e is not None else 0.0)
        gt_mos.append(float(row['mos']))

        # Save row immediately after each image — crash-safe
        with open(csv_path, 'a', newline='') as f:
            writer = csv.writer(f)
            writer.writerow([
                row['img_path'],
                zsl_s   if zsl_s is not None else '',
                zsl_e   if zsl_e is not None else '',
                q,
                float(row['mos']),
            ])
    else:
        print(f"Skipped image due to error: {row['img_path']}")
        # Save failed row too — useful for debugging
        with open(csv_path, 'a', newline='') as f:
            writer = csv.writer(f)
            writer.writerow([row['img_path'], '', '', '', float(row['mos'])])

# Convert to numpy arrays for correlation metrics
vilar_scores        = np.array(vilar_scores)
zsl_single_scores   = np.array(zsl_single_scores)
zsl_ensemble_scores = np.array(zsl_ensemble_scores)
gt_mos              = np.array(gt_mos)

print(f"\n--- RESULTS ---")
print(f"Successfully processed: {len(gt_mos):,} images")
print(f"Saved to:               {csv_path}")
```

```python
from concurrent.futures import ThreadPoolExecutor, as_completed
import threading

# Thread-safe CSV writer
csv_lock = threading.Lock()

def process_single_image(row):
    """Process one image — called in parallel."""
    q, zsl_s, zsl_e = run_vila_predict(row['img_path'], CKPT_VILAR, SPM_PATH)

    if q is not None:
        # Thread-safe write to CSV
        with csv_lock:
            with open(csv_path, 'a', newline='') as f:
                writer = csv.writer(f)
                writer.writerow([
                    row['img_path'],
                    zsl_s if zsl_s is not None else '',
                    zsl_e if zsl_e is not None else '',
                    q,
                    float(row['mos']),
                ])
        return q, zsl_s, zsl_e, float(row['mos'])
    else:
        print(f"Skipped: {row['img_path']}")
        return None
```

```python
# --- PARALLEL EVALUATION ---
MAX_WORKERS = 4  # Liczba równoległych procesów
                 # Na Colab zalecane: 2-4 (więcej = OOM risk)

vilar_scores        = []
zsl_single_scores   = []
zsl_ensemble_scores = []
gt_mos              = []

rows = [row for _, row in df_eval.iterrows()]

with ThreadPoolExecutor(max_workers=MAX_WORKERS) as executor:
    # Submit all jobs
    futures = {
        executor.submit(process_single_image, row): row
        for row in rows
    }

    # Collect results as they complete
    for future in tqdm(
        as_completed(futures),
        total=len(futures),
        desc=f'VILA Evaluation (workers={MAX_WORKERS})'
    ):
        result = future.result()
        if result is not None:
            q, zsl_s, zsl_e, mos = result
            vilar_scores.append(q)
            zsl_single_scores.append(zsl_s   if zsl_s is not None else 0.0)
            zsl_ensemble_scores.append(zsl_e if zsl_e is not None else 0.0)
            gt_mos.append(mos)

vilar_scores        = np.array(vilar_scores)
zsl_single_scores   = np.array(zsl_single_scores)
zsl_ensemble_scores = np.array(zsl_ensemble_scores)
gt_mos              = np.array(gt_mos)

print(f"\n--- RESULTS ---")
print(f"Successfully processed: {len(gt_mos):,} images")
print(f"Saved to:               {csv_path}")
```

```python
def metrics(pred, gt, name, srcc_p, plcc_p):
    srcc, _ = spearmanr(pred, gt)
    plcc, _ = pearsonr(pred,  gt)
    return {
        'Model':        name,
        'SRCC (ours)':  round(srcc, 3),
        'SRCC (paper)': srcc_p,
        'ΔSRCC':        round(srcc - srcc_p, 3),
        'PLCC (ours)':  round(plcc, 3),
        'PLCC (paper)': plcc_p,
        'ΔPLCC':        round(plcc - plcc_p, 3),
    }

results = pd.DataFrame([
    metrics(zsl_scores,   gt_mos, 'VILA-P (ZSL)', 0.657, 0.663),
    metrics(vilar_scores, gt_mos, 'VILA-R',        0.774, 0.774),
])

print("\n===== REPRODUKCJA TABELI 1 =====\n")
print(results.to_string(index=False))
```
